```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
crime=pd.read_csv('D:\Assignments\Clustering\crime_data.csv')
print(crime)
```

```python
#normalize the data
def norm_func(i):
    x = (i-i.min())	/	(i.max()	-	i.min())
    return (x)
```

```python
df_norm=norm_func(crime.iloc[:,1:])#considering the data from 1st column
df_norm
```

```python
from scipy.cluster.hierarchy import linkage
import scipy.cluster.hierarchy as sch
#using ecludiean distance method for finding distance matrix #converting it into array
p = linkage(df_norm,method='complete',metric='euclidean')
p
```

```python
#visualizing the dendogram ,distance matrix
plt.figure(figsize=(20,10));plt.title('Hierarchical Clustering Dendrogram');plt.xlabel('Index');plt.ylabel('Distance')
sch.dendrogram(
    p,
    leaf_rotation=2.,  # rotates the x axis labels
    leaf_font_size=6.,  # font size for the x axis labels
)
plt.show()
```

```python
from	sklearn.cluster	import	AgglomerativeClustering 
#aplying on model
h_clust=AgglomerativeClustering(n_clusters=5,linkage="complete",affinity="euclidean").fit(df_norm)
cluster_labels=pd.Series(h_clust.labels_)
cluster_labels #labels the clusters
crime['clust']=cluster_labels #assigning the labels to a new column 'clust'
#getting aggregate mean of each cluster
crime=crime.iloc[:,].groupby(crime.clust).mean()
crime
```
